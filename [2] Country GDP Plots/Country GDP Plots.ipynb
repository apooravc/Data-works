## Country GDP Plots

**Description** 
<br>
Plotting GDP of 10 European countries with time on a yearly basis during 1960-2017.

**References**
<br>
1) [Country GDP series data from World Development Indicators - World Bank](http://datatopics.worldbank.org/world-development-indicators/themes/economy.html)

```python
import pandas as pd

country_gdp_df = pd.read_csv("API_NY.GDP.MKTP.CD_DS2_en_csv_v2_10224782.csv") 
country_gdp_df.head()
```

```python
country_index_list = [35, 53, 56, 68, 75, 79, 114, 174, 200, 221] # index of selected countries to plot
AllCountry_gdp_list = [] # list of GDP series (during 1960-2017) list of all countries
AllCountry_name_list = [] # list of name of all countries
for index, row in country_gdp_df.iterrows():
    if index in country_index_list:
        country_gdp_byYear = row["1960":"2017"].tolist() # GDP series (during 1960-2017) list of a single country
        AllCountry_gdp_list.append(country_gdp_byYear)
        country_name = row["Country Name"] # name of a single country
        AllCountry_name_list.append(country_name)
```

```python
year_list = list(country_gdp_df) # list of DataFrame column headers
year_list = year_list[4:-1] # extracting list of 1960-2017 headers
year_list = [int(i) for i in year_list]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
for country_gdp_byYear in AllCountry_gdp_list:
    plt.plot(year_list, country_gdp_byYear)
    
plt.xticks([year for year in range(1960, 2018, 3)])
plt.rcParams['figure.figsize'] = (15.0, 8.0) # for resizing plot
plt.legend(AllCountry_name_list, loc="best")
plt.title("Line Plot of GDP of 10 European countries with time")
plt.xlabel("Time (year)")
plt.ylabel("GDP (current US$)")
plt.show()
```
