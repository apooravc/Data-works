## Wikipedia word cloud

**Description** 
<br>
Creating a word cloud using text of a Wikipedia page. A word cloud is a graphical representation of frequency distribution of different words in a text.

```python
# first, get text data using python library for Wikipedia
import wikipedia

title = "Python (programming language)" # title of Wikipedia page for word cloud

# watch out for 2 possible errors: 
# 1) DisambiguationError - when given title refers to multiple pages
# 2) PageError - when page of given title doesn't exist
try: 
    wiki_page = wikipedia.page(title) # wikipedia.page is used to access data of a Wikipedia page using its title
except wikipedia.exceptions.DisambiguationError as e: # when title = "Python"
    print('"' + title + '"', "may refer to any of ", e.options)
except wikipedia.exceptions.PageError as e: # when title = "zyvx"
    print('"' + title + '"', "didn't match any existing page.")
    
wiki_text = wiki_page.content # returns text content of page in string format
```

```python
# second, generate word cloud of a particular shape using a mask and text data
from wordcloud import WordCloud
import numpy as np
from PIL import Image

mask = np.array(Image.open("cloud.png")) # using a mask to create word cloud of "cloud" shape
wiki_wordcloud = WordCloud(background_color = "white", max_words = 250, mask = mask).generate(wiki_text) # generating word cloud
wiki_wordcloud.to_file("wordcloud.jpeg") # exporting generated word cloud to an image file
```

```python
# third, display generated image using matplotlib
import matplotlib.pyplot as plt
plt.imshow(wiki_wordcloud, interpolation = "bilinear")
plt.axis("off")
```
