## Country GDP Plots

**Description** 
<br>
Plotting 3 parameters of 10 European countries with time on a yearly basis during 1960-2017. 
<br>
These 3 parameters being - GDP, Population and GDP per Capita.

**References**
<br>
1) [Country GDP (current USD) series (1960-2017) data from World Development Indicators - World Bank](http://datatopics.worldbank.org/world-development-indicators/themes/economy.html)
<br>
2) [Country Population (total) series (1960-2017) data from World Development Indicators - World Bank](http://datatopics.worldbank.org/world-development-indicators/themes/people.html)

```python
country_name_list = ['Switzerland', 'Germany', 'Denmark', 'Spain', 'France', 'United Kingdom', 'Italy', 'Netherlands', 'Russian Federation', 'Sweden'] # index of selected countries to plot
```

```python
import pandas as pd

# first DataFrame - GDP series 
country_gdp_df = pd.read_csv("API_NY.GDP.MKTP.CD_DS2_en_csv_v2_10224782.csv", index_col=0) # selecting column "Country Name" as index
country_gdp_df.head()
```

```python
# second DataFrame - Population series 
country_pop_df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_10224786.csv", index_col=0) # selecting column "Country Name" as index
country_pop_df.head()
```

```python
# subset selection from first DataFrame
country_gdp_df = country_gdp_df.loc[country_name_list, :] # selecting rows having "Country Name" in country_name_list
country_gdp_df = country_gdp_df.loc[:, "1960":"2017"] # selecting columns having labels from 1960 to 2017
country_gdp_df
```

```python
# subset selection from second DataFrame
country_pop_df = country_pop_df.loc[country_name_list, :] # selecting rows having "Country Name" in country_name_list
country_pop_df = country_pop_df.loc[:, "1960":"2017"] # selecting columns having labels from 1960 to 2017
country_pop_df
```

```python
# preparing data for X-axis in plots - Time (year)
year_list = [year for year in range(1960, 2018)]
```

```python
# preparing data for first plot - GDP vs. Time
AllCountry_gdp_list = [] # list of GDP series (during 1960-2017) list of selected countries
for index, row in country_gdp_df.iterrows():
    country_gdp_byYear = row.tolist() # GDP series (during 1960-2017) list of a single country
    AllCountry_gdp_list.append(country_gdp_byYear)
```

```python
# preparing data for second plot - Population vs. Time
AllCountry_pop_list = [] # list of population series (during 1960-2017) list of selected countries
for index, row in country_pop_df.iterrows():
    country_pop_byYear = row.tolist() # population series (during 1960-2017) list of single country
    AllCountry_pop_list.append(country_pop_byYear)
```

```python
# preparing data for third plot - GDP per Capita vs. Time
country_gdpPerCap_df = country_gdp_df.div(country_pop_df, axis='columns', fill_value=None) # dividing GDP by Population for each
country_gdpPerCap_df
```

```python
# preparing data for third plot - GDP per Capita vs. Time
AllCountry_gdpPerCap_list = [] # list of GDP per Capita series (during 1960-2017) list of selected countries
for index, row in country_gdpPerCap_df.iterrows():
    country_gdpPerCap_byYear = row.tolist() # GDP per Capita series (during 1960-2017) list of single country
    AllCountry_gdpPerCap_list.append(country_gdpPerCap_byYear)
```

```python
# Plotting 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# first plot
for country_gdp_byYear in AllCountry_gdp_list:
    plt.plot(year_list, country_gdp_byYear)
    
plt.xticks([year for year in range(1960, 2018, 3)])
plt.ticklabel_format(style="plain", axis="y")
plt.rcParams['figure.figsize'] = (15.0, 8.0) # for resizing plot
plt.legend(country_name_list, loc="best")
plt.title("Line Plot of GDP of 10 European countries with time")
plt.xlabel("Time (year)")
plt.ylabel("GDP (current US$)")
plt.show()
```

```python
# second plot
for country_pop_byYear in AllCountry_pop_list:
    plt.plot(year_list, country_pop_byYear)
    
plt.xticks([year for year in range(1960, 2018, 3)])
plt.ticklabel_format(style="plain", axis="y")
plt.rcParams['figure.figsize'] = (15.0, 8.0) # for resizing plot
plt.legend(country_name_list, loc="best")
plt.title("Line Plot of Population of 10 European countries with time")
plt.xlabel("Time (year)")
plt.ylabel("Population (total)")
plt.show()
```

```python
# third plot
for country_gdpPerCap_byYear in AllCountry_gdpPerCap_list:
    plt.plot(year_list, country_gdpPerCap_byYear)

plt.xticks([year for year in range(1960, 2018, 3)])
plt.ticklabel_format(style="plain", axis="y")
plt.rcParams['figure.figsize'] = (15.0, 8.0) # for resizing plot
plt.legend(country_name_list, loc="best")
plt.title("Line Plot of GDP per Capita of 10 European countries with time")
plt.xlabel("Time (year)")
plt.ylabel("GDP per Capita (USD)")
plt.show()
```
